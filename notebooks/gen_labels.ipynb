```python
from pathlib import Path
import json

input_dir = Path("/home/max/Documents/RA/egowholemocap/notebooks/undistorted")  

output_json = Path("/home/max/Documents/RA/ovmono3d/datasets/insta360/labels.json")

labels = ["cup", "book"]

files = [p for p in input_dir.iterdir() if p.is_file()]

mapping = {p.stem: labels for p in sorted(files, key=lambda p: int(p.stem))}

output_json.write_text(json.dumps(mapping, indent=2))
print(f"Wrote {len(mapping)} entries to {output_json}")
```

```python
# reorder pickle elements from alphabetical to numerical order
import pickle
from pathlib import Path

pickle_file = Path("/home/max/Documents/RA/ovmono3d/output/insta360_unprocessed/detections.pkl") 
folder_with_numeric_files = Path("/home/max/Documents/RA/egowholemocap/notebooks/unprocessed") 

with open(pickle_file, 'rb') as f:
    values = pickle.load(f)

print(f"Loaded {len(values)} values from pickle file")

files = sorted([p for p in folder_with_numeric_files.iterdir() if p.is_file()])

# Create alphabetical order (how files were originally processed)
files_alphabetical = sorted(files, key=lambda p: p.stem)

# Create numerical order (how we want them)
files_numerical = sorted(files, key=lambda p: int(p.stem))

mapping = {}
for i, file in enumerate(files_alphabetical):
    numerical_idx = files_numerical.index(file)
    mapping[i] = numerical_idx

print(f"\nProcessed {len(files)} files")
print(f"Example mapping (alphabetical idx -> numerical idx): {dict(list(mapping.items())[:5])}")

reordered_values = [None] * len(values)
for old_idx, new_idx in mapping.items():
    reordered_values[new_idx] = values[old_idx]

output_pickle = pickle_file.parent / f"{pickle_file.stem}_reordered.pkl"
with open(output_pickle, 'wb') as f:
    pickle.dump(reordered_values, f)

print(f"\nSaved reordered values to: {output_pickle}")
```
